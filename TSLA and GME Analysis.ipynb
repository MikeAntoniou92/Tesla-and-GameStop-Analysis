```python
#Libraries Used"
import yfinance as yf
import pandas as pd
import requests as rq
from bs4 import BeautifulSoup
```

Question 1 - Extracting Tesla Stock Data Using yfinance

```python
TSLA = yf.Ticker("TSLA")
TSLA_data = TSLA.history(period="max")
TSLA_data.reset_index(inplace=True)
#TSLA_data.head(5)
```

Question 2 - Extracting Tesla Revenue Data Using Webscraping

```python
TSLA_url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/labs/project/revenue.htm"
TSLA_htmlRevenues  = rq.get(TSLA_url).text
soup = BeautifulSoup(TSLA_htmlRevenues, 'html5lib')
TSLA_Revenues = pd.DataFrame(columns=["Year", "Revenue in millions"])
for row in soup.find("tbody").find_all('tr'):
    col = row.find_all("td")
    Year=col[0].text
    Revenue=col[1].text
    TSLA_Revenues= pd.concat([TSLA_Revenues, pd.DataFrame([{'Year':Year, 'Revenue in millions':Revenue}])], ignore_index=True)
TSLA_Revenues["Revenue in millions"] = TSLA_Revenues["Revenue in millions"].str.replace('$',"")
TSLA_Revenues["Revenue in millions"] = TSLA_Revenues["Revenue in millions"].str.replace(',',"")
#TSLA_Revenues.tail(5)
```

Question 3 - Extracting GameStop Stock Data Using yfinance

```python
GME = yf.Ticker("GME")
GME_data = GME.history(period="max")
GME_data.reset_index(inplace=True)
GME_data.head(5)
```

Question 4 - Extracting GameStop Revenue Data Using Webscraping

```python
gme_url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/labs/project/stock.html"
gme_htmlrevenues  = rq.get(gme_url).text
soup = BeautifulSoup(gme_htmlrevenues, 'html5lib')
GME_Revenues = pd.DataFrame(columns=["Year", "Revenue in millions"])
for row in soup.find("tbody").find_all('tr'):
    col = row.find_all("td")
    Year=col[0].text
    Revenue=col[1].text
    GME_Revenues= pd.concat([GME_Revenues, pd.DataFrame([{'Year':Year, 'Revenue in millions':Revenue}])], ignore_index=True)
GME_Revenues["Revenue in millions"] = GME_Revenues["Revenue in millions"].str.replace('$',"")
GME_Revenues["Revenue in millions"] = GME_Revenues["Revenue in millions"].str.replace(',',"")
GME_Revenues.tail(5)
```

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots
def make_graph(stock_data, revenue_data, stock):
    fig = make_subplots(rows=2, cols=1, shared_xaxes=True, subplot_titles=("Historical Share Price", "Historical Revenue"), vertical_spacing = .5)
    stock_data_specific = stock_data["Date"]
    revenue_data_specific = revenue_data["Year"]
    fig.add_trace(go.Scatter(x=stock_data_specific, y=stock_data["Close"].astype("float"), name="Share Price"), row=1, col=1)
    fig.add_trace(go.Scatter(x=revenue_data_specific, y=revenue_data["Revenue in millions"].astype("float"), name="Revenue"), row=2, col=1)
    fig.update_xaxes(title_text="Date", row=1, col=1)
    fig.update_xaxes(title_text="Date", row=2, col=1)
    fig.update_yaxes(title_text="Price ($US)", row=1, col=1)
    fig.update_yaxes(title_text="Revenue ($US Millions)", row=2, col=1)
    fig.update_layout(showlegend=False,
    height=600,
    title=stock,
    xaxis_rangeslider_visible=True)
    fig.show()
```

Question 5 - Tesla Stock and Revenue Dashboard

```python
make_graph(TSLA_data, TSLA_Revenues, 'GameStop Stock Price/Revenue Dashboard')
```

Question 6 - GameStop Stock and Revenue Dashboard

```python
make_graph(GME_data, GME_Revenues, 'GameStop Stock Price/Revenue Dashboard')
```
